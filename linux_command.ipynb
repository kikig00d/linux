#ls 명령어 설명
ls (list segments) 명령어는 리눅스/유닉스 환경에서 디렉토리 내부의 파일과 하위 디렉토리 목록을 화면에 출력하는 핵심 명령어입니다. 기본적으로 현재 디렉토리의 내용을 보여주며, 다양한 옵션을 통해 상세 정보, 숨김 파일, 정렬 방식 등을 제어할 수 있습니다.
#주요 ls 명령어 옵션:
ls -l: 상세 정보(권한, 소유자, 크기, 수정 시간)를 롱 포맷으로 표시.
ls -a: 숨김 파일(.)을 포함한 모든 파일 표시.
ls -al (또는 ls -la): 숨김 파일을 포함하여 모든 파일의 상세 정보를 표시 (가장 많이 사용).
ls -lh: 파일 크기를 읽기 쉬운 단위(KB, MB, GB)로 표시.
ls -t: 파일의 수정 시간 순서대로 정렬.
ls -r: 목록을 역순으로 표시.
ls -R: 하위 디렉토리의 내용까지 재귀적으로 표시

#pwd (Print Working Directory)
현재 내 위치를 알려주는 명령어. 현재 위치 확인.
*   의미:현재 작업 중인 디렉터리의 전체 경로를 화면에 출력.
*   사용법: 단순히 pwd라고 입력하고 엔터
*   출력 예시: /home/user/documents

#ls(List)
이 폴더의 구성요소 확인 명령어. 주변 오브젝트 탐색.
* 의미: 현재 디렉터리에 있는 파일과 하위 디렉터리의 목록을 나타냄
* 자주 쓰는 옵션:
* ls -l: 파일의 권한, 소유자, 크기, 수정 날짜 등 상세 정보 표시
* ls -a: 숨겨진 파일까지 모두 표시
* ls -al: 두 기능을 합쳐서 모든 파일 상세 표시
#cd (Change Directory)
다른 폴더로 이동 명령어. 특정 좌표/구역 이동
* 의미: 작업 디렉터리를 변경
* 사용법: cd[이동할 경로]
* 주요 이동 공식:
* cd ..: 한 단계 위 상위 폴더로 이동
* cd /: 최상위 디렉터리로 이동
* cd ~:현재 사용자의 홈 디렉터리(내 폴더)로 이동
* cd -:바로 직전의 디렉터리로 회귀

# chmod(Change Mode) 파일/디렉터리 사용 권한 설정
read, write, excute로 rwx, 000으로 되어 있으며,
기본적으로 754와 같이 입력해서 쓴다.
* 사용법: chmod [숫자] 파일명
* chmod 700 secret.txt 해당 텍스트의 사용권한은 사용자만 있다.
* 맨 앞은 소유자 권한. 7 = 111, rwx 모두 가능
* 중간은 그룹 권한. 5 = 101, r_x만 가능
* 맨 뒤는 기타 권한, 방문자 권한. 100, r__만 가능

#chown (Change Owner) 파일/디렉터리 소유권 이전
chown kim aaa.txt
킴에게 aaa 텍스트 파일의 소유권을 준다.
* 사용법: sudo chown [유저이름]:[그룹이름] [파일명] 유저나 그룹은 둘 중 하나 혹은 둘 다 정할 수 있다.
* /etc나 /bin은 소유자를 일반 유저로 바꿀 수 없고 강제하면 운영체제가 망가질 수 있다. 그것은 모두 root의 소유.
* 실행되어 계속 돌아가야하는 파일에 chmod -x 와 같이 실행 권한 뺏기를 하면 그 즉시 서비스가 멈춘다.
