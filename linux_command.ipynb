#ls 명령어 설명
ls (list segments) 명령어는 리눅스/유닉스 환경에서 디렉토리 내부의 파일과 하위 디렉토리 목록을 화면에 출력하는 핵심 명령어입니다. 기본적으로 현재 디렉토리의 내용을 보여주며, 다양한 옵션을 통해 상세 정보, 숨김 파일, 정렬 방식 등을 제어할 수 있습니다.
#주요 ls 명령어 옵션:
ls -l: 상세 정보(권한, 소유자, 크기, 수정 시간)를 롱 포맷으로 표시.
ls -a: 숨김 파일(.)을 포함한 모든 파일 표시.
ls -al (또는 ls -la): 숨김 파일을 포함하여 모든 파일의 상세 정보를 표시 (가장 많이 사용).
ls -lh: 파일 크기를 읽기 쉬운 단위(KB, MB, GB)로 표시.
ls -t: 파일의 수정 시간 순서대로 정렬.
ls -r: 목록을 역순으로 표시.
ls -R: 하위 디렉토리의 내용까지 재귀적으로 표시

# chmod(Change Mode) 파일/디렉터리 사용 권한 설정
read, write, excute로 rwx, 000으로 되어 있으며,
기본적으로 754와 같이 입력해서 쓴다.
* 사용법: chmod [숫자] 파일명
* chmod 700 secret.txt 해당 텍스트의 사용권한은 사용자만 있다.
* 맨 앞은 소유자 권한. 7 = 111, rwx 모두 가능
* 중간은 그룹 권한. 5 = 101, r_x만 가능
* 맨 뒤는 기타 권한, 방문자 권한. 100, r__만 가능

#chown (Change Owner) 파일/디렉터리 소유권 이전
chown kim aaa.txt
킴에게 aaa 텍스트 파일의 소유권을 준다.
* 사용법: sudo chown [유저이름]:[그룹이름] [파일명] 유저나 그룹은 둘 중 하나 혹은 둘 다 정할 수 있다.
* /etc나 /bin은 소유자를 일반 유저로 바꿀 수 없고 강제하면 운영체제가 망가질 수 있다. 그것은 모두 root의 소유.
* 실행되어 계속 돌아가야하는 파일에 chmod -x 와 같이 실행 권한 뺏기를 하면 그 즉시 서비스가 멈춘다.

#find
이름으로 찾기, 타입으로 찾기, 크기로 찾기, 시간으로 찾기, 찾은 후 바로 실행하기 등의 기능이 있다. 마지막으로 찾아서 지우기.

#이름으로 찾기(-name)
대소문자 구분하기 싫으면 -iname 으로 쓴다.
* 현재 폴더에서 이름으로 찾기

find . -name "ubuntu.txt"

* 특정 폴더에서 확장자로 찾기

find /home -name "*.jpg"

#타입으로 찾기(-type)
찾는 대상이 파일인지 폴더인지 구분할 때 쓴다

* 디렉터리(폴더)만 찾기:

find . -type d -name "my_project"

* 일반 파일만 찾기:

find . -type f -name "config*"

#크기로 찾기 (-size)
용량이 너무 큰 파일을 찾아서 지우고 싶을 때 유용합니다.

* 100MB 이상인 파일 찾기:

find . -size +100M

* 1GB 이상인 파일 찾기:

find . -size +1G

* 10k 이하인 파일 찾기:

find . -size -10k

#시간으로 찾기 (-mtime)
"어제 수정한 파일이 뭐였지?" 싶을 때 사용하는 시간 탐지 기능입니다.

* 최근 24시간 내에 수정된 파일:

find . -mtime -1

* 7일보다 오래전에 수정된 파일:

find . -mtime +7

#찾은 후 바로 실행하기 (-exec) (고급)
이게 find의 진정한 강력함입니다. 파일을 찾자마자 다른 명령어를 바로 연결할 수 있습니다.

* 찾은 파일들 전부 삭제하기 (주의!):

find . -name "*.log" -exec rm -f {} \;

{}는 찾은 파일의 이름을 뜻하고, \;는 명령어의 끝을 알립니다.

* 찾은 파일들의 내용물 확인하기:

find . -name "memo.txt" -exec cat {} \;

#systemctl
시스템의 모든 서비스(백그라운드 프로그램)을 관리하는 명령어.
해당 명령어가 다루는 대상을 보통 유닛(Unit)이라고 부른다. 대표적인 게 서비스.

#가장 많이 쓰는 실전 명령어 (서비스 제어)
게임기로 치면 전원을 켜고 끄거나 상태창을 확인하는 것과 같습니다.

* 상태 확인: systemctl status [서비스명]

예: systemctl status nginx (지금 웹 서버가 잘 돌아가고 있나?)

* 시작: systemctl start [서비스명]

프로그램을 지금 당장 실행합니다.

* 중지: systemctl stop [서비스명]

프로그램을 즉시 종료합니다.

* 재시작: systemctl restart [서비스명]

설정을 바꾼 후 껐다가 다시 켤 때 사용합니다.

#부팅 시 자동 실행 설정 (Enable / Disable)
컴퓨터를 껐다 켜도 자동으로 프로그램이 실행되게 할지 정하는 중요한 설정입니다. start와는 별개입니다.

* 자동 실행 등록: systemctl enable [서비스명]

"다음에 컴퓨터 켜질 때 이 프로그램도 같이 켜줘."

* 자동 실행 해제: systemctl disable [서비스명]

"이제 컴퓨터 켜질 때 이 프로그램은 안 켜도 돼."

#전체 목록 확인하기
내 컴퓨터에서 어떤 서비스들이 돌아가고 있는지 한눈에 보고 싶을 때 사용합니다.

* 실행 중인 모든 서비스 보기:

systemctl list-units --type=service

* 문제가 생긴(실패한) 서비스만 보기:

systemctl --failed

#시스템 전체 제어 (전원)
systemctl은 개별 프로그램뿐만 아니라 시스템 전체의 전원도 관리합니다.

* 재부팅: systemctl reboot

* 전원 끄기: systemctl poweroff

#grep (Global Regular Expression Print)
내가 원하는 특정 문자열(단어)이 포함된 줄만 찾아주는 명령어.
돋보기나 탐색기 같은 역할.
* 기본 사용법

grep [찾을단어] [파일명]

* 옵션

-i(Ignore case): 대소문자 구분하지 않음.

-v(Invert match): 특정 단어가 포함되지 않은 줄만 표시.

-r(Recursive): 현재 폴더뿐만 아닌 하위 폴더의 모든 파일을 다 확인.

-n(Line number): 찾은 단어가 몇 번째 줄에 이는지 번호를 붙인다.

#프로세스 관리 명령어
ps, -ef, grep, kill, &/검색, 검색, 종료, 잠수
* ps -ef 실행 중인 프로세스 추적 명령어

-e: Everything 모든

-f: Full format 사용자 ID, 부모 프로세스 번호(PPID), 시작 시간 등 상세 정보 표시

* grep 원하는 단어, 문장을 검색 명령어

활용: ps -ef | grep [이름] 형식을 쓴다

* kill 프로세스를 종료하는 명령어

사용법 : kill [옵션] [PID]

-9 옵션: kill -9와 같이 쓰며 강제 종료 신호다.
kill -9 1234 와 같이 쓴다. 1234는 프로그램 번호.

* &(앰퍼샌드) 백그라운드에 잠수된 체로 실행하게 하는 특수 기호.

사용 목적: 터미널에서 프로그램(gedit 등)을 실행하면 그 프로그램이 종료될 때까지 다른 명령어를 입력할 수 없다. &를 붙이면 백그라운드로 이동해 터미널이 자유로워진다.

예시: gedit&
